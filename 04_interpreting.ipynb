#### Category Transform

```python
print(f'변환 전 메모리 사용량 : {df.memory_usage().sum():>10,.0f}')
```

```python
filt_object = df.dtypes == object
object_features = list(df.dtypes[filt_object].index)      # 문자형 변수
for var in object_features:
  df[var] = df[var].astype('category')
```

```python
print(f'변환 후 메모리 사용량 : {df.memory_usage().sum():>10,.0f}')
```

```python
# 카테고리 순서 지정
## education
education_cat = CategoricalDtype(categories=\
              ['Under Elementary School', 'Middle School', 'High School', 'College', 'University', 'Master, Doctor'], 
              ordered=True)
df.education = df.education.astype(education_cat)

## work_class
work_class_cat = CategoricalDtype(categories=['Day Worker', 'Temporary Worker', 'Regular Worker'], ordered=True)
df.work_class = df.work_class.astype(work_class_cat)

## jobin
jobin_list = ['jobin_1', 'jobin_2', 'jobin_3', 'jobin_4', 'jobin_5', 'jobin_6', 'jobin_7', 'jobin_8', 'jobin_9']
jobin_cat = CategoricalDtype(categories=['Very Bad', 'Bad', 'Normal', 'Good', 'Very Good'], ordered=True)

for var in jobin_list:
  df[var] = df[var].astype(jobin_cat)

## 'skill_2'
skill_2_cat = CategoricalDtype(categories=\
                                 ['Very Low', 'Low', 'Suit', 'High', 'Very High'], 
                                 ordered=True)
df.skill_2 = df.skill_2.astype(skill_2_cat)

## skill_3
skill_3_cat = CategoricalDtype(categories=\
                                 ['Very Useless', 'Useless', 'Useful', 'Very Useful'], 
                                 ordered=True)
df.skill_3 = df.skill_3.astype(skill_3_cat)
```

```python
print(f'변환 후 메모리 사용량 : {df.memory_usage().sum():>10,.0f}')
```

#### Split Data

```python
# target
filt_com = df['work_industry'] == 'Computing Data'
filt_ret = df['work_industry'] == 'Retail'
filt_bld = df['work_industry'] == 'Building'

# 전체 데이터
X_all = df.drop(target, axis=1)
y_all = df[target]
all_X_train, all_X_test, all_y_train, all_y_test = train_test_split(X_all, y_all, test_size=0.2, random_state=rand_seed)
print(f'전체 : {all_X_train.shape} / {all_X_test.shape} / {all_y_train.shape} / {all_y_test.shape}')

# 컴퓨터 직군 분할
X_com = df[filt_com].drop(target, axis=1)
y_com = df[filt_com][target]
com_X_train, com_X_test, com_y_train, com_y_test = train_test_split(X_com, y_com, test_size=0.2, random_state=rand_seed)
print(f'컴퓨터 직군 : {com_X_train.shape} / {com_X_test.shape} / {com_y_train.shape} / {com_y_test.shape}')

# 소매업 직군 분할
X_ret = df[filt_ret].drop(target, axis=1)
y_ret = df[filt_ret][target]
ret_X_train, ret_X_test, ret_y_train, ret_y_test = train_test_split(X_ret, y_ret, test_size=0.2, random_state=rand_seed)
print(f'소매업 직군 : {ret_X_train.shape} / {ret_X_test.shape} / {ret_y_train.shape} / {ret_y_test.shape}')

# 건설업 직군 분할
X_bld = df[filt_bld].drop(target, axis=1)
y_bld = df[filt_bld][target]
bld_X_train, bld_X_test, bld_y_train, bld_y_test = train_test_split(X_bld, y_bld, test_size=0.2, random_state=rand_seed)
print(f'건설업 직군 : {bld_X_train.shape} / {bld_X_test.shape} / {bld_y_train.shape} / {bld_y_test.shape}')
```

```python
def print_score(model, X_train, y_train, cv_n=3) :
  """ 학습 된 모델에서 평가지표를 반환하는 함수 """

  np.random.seed(rand_seed)
  
  # 타겟 추정
  y_train_pred      = model.predict(X_train)
  y_train_proba     = model.predict_proba(X_train)[:, 1]

  train_accuracy    = accuracy_score  (y_train, y_train_pred)
  train_f1          = f1_score        (y_train, y_train_pred)
  train_roc_auc     = roc_auc_score   (y_train, y_train_proba)

  val_accuracy_cv      = cross_val_score(model, X_train, y_train, scoring='accuracy', cv=cv_n)
  val_f1_cv            = cross_val_score(model, X_train, y_train, scoring='f1', cv=cv_n)
  val_roc_auc_cv       = cross_val_score(model, X_train, y_train, scoring='roc_auc', cv=cv_n)


  result_df = pd.DataFrame({'train' : [train_accuracy, train_f1, train_roc_auc],
                            'val_mean'   : [val_accuracy_cv.mean(), val_f1_cv.mean(), val_roc_auc_cv.mean()],
                            'val_std'   : [val_accuracy_cv.std(), val_f1_cv.std(), val_roc_auc_cv.std()],
                            }, index=['Accuracy','F1','AUC'])
  
  print(result_df.T)

  return result_df.T
```

```python
def rf_pipe_fit(X_train, y_train):
    """ Random Forest Classifier 파이프를 만들고 모델에 학습하는 함수 """
    rf_pipe = make_pipeline(
                    OrdinalEncoder(),
                    RandomForestClassifier(random_state       = rand_seed
                                          , class_weight      = 'balanced'
                                          , max_depth         = int(best_params_read.loc['rf_max_depth'][0])
                                          , min_samples_leaf  = int(best_params_read.loc['rf_min_samples_leaf'][0])
                                          ),
                    )
    return rf_pipe.fit(X_train, y_train)

def xg_pipe_fit(X_train, y_train):
    """ XGBClassifier 파이프를 만들고 모델에 학습하는 함수 """
    xg_pipe = make_pipeline(
                    OrdinalEncoder(),
                    XGBClassifier(random_state          = rand_seed
                                  , scale_pos_weight    = y_train.value_counts()[0] / y_train.value_counts()[1]
                                  , max_depth           = int(best_params_read.loc['xg_max_depth'][0])
                                  , min_child_weight    = int(best_params_read.loc['xg_min_child_weight'][0])
                                  ),
                    )
    return xg_pipe.fit(X_train, y_train)
```
