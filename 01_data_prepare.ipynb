```python
def feature_engineering(df):
  """ 특성 공학 함수 """

  ############################################################################
  # sex : 성별
  sex_map = {1:'Male', 2:'Female'}
  df.sex = df.sex.map(sex_map)

  ############################################################################
  # education : 교육 수준
  edu_map = {
      1       : 'Under Elementary School', 
      2       : 'Under Elementary School', 
      3       : 'Under Elementary School', 
      4       : 'Middle School', 
      5       : 'High School', 
      6       : 'College', 
      7       : 'University', 
      8       : 'Master, Doctor', 
      9       : 'Master, Doctor', 
      -1      : 'Under Elementary School', 
      np.nan  : 'Under Elementary School', 
      # -1      : np.nan, 
      }
  df.education = df.education.map(edu_map)

  ############################################################################
  # region : 거주지역 수도권, 비수도권 이진화
  filt_region = (df['region'] == 1) | \
                (df['region'] == 5) | \
                (df['region'] == 8)
  df['seoul'] = 'No Seoul'
  df.loc[df[filt_region].index, 'seoul'] = 'Seoul'

  ############################################################################
  # work_kind : 표준직업분류 5차
  work_kind_map = {
      11      : 'High Official', 
      12      : 'High Official', 
      13      : 'High Official', 
      21      : 'High Official', 
      22      : 'High Official', 
      23      : 'High Official', 
      24      : 'High Official', 
      30      : 'High Official', 
      111     : 'Expert', 
      112     : 'Expert', 
      113     : 'Expert', 
      120     : 'Expert', 
      131     : 'Expert', 
      132     : 'Expert', 
      133     : 'Expert', 
      134     : 'Expert', 
      134     : 'Expert', 
      141     : 'Expert', 
      142     : 'Expert', 
      143     : 'Expert', 
      144     : 'Expert', 
      145     : 'Expert', 
      151     : 'Expert', 
      152     : 'Expert', 
      153     : 'Expert', 
      154     : 'Expert', 
      155     : 'Expert', 
      156     : 'Expert', 
      157     : 'Expert', 
      161     : 'Expert', 
      162     : 'Expert', 
      163     : 'Expert', 
      164     : 'Expert', 
      165     : 'Expert', 
      171     : 'Expert', 
      172     : 'Expert', 
      173     : 'Expert', 
      181     : 'Expert', 
      182     : 'Expert', 
      183     : 'Expert', 
      184     : 'Expert', 
      211     : 'Semi Expert', 
      212     : 'Semi Expert', 
      213     : 'Semi Expert', 
      220     : 'Semi Expert', 
      231     : 'Semi Expert', 
      232     : 'Semi Expert', 
      233     : 'Semi Expert', 
      234     : 'Semi Expert', 
      235     : 'Semi Expert', 
      236     : 'Semi Expert', 
      237     : 'Semi Expert', 
      238     : 'Semi Expert', 
      241     : 'Semi Expert', 
      242     : 'Semi Expert', 
      243     : 'Semi Expert', 
      251     : 'Semi Expert', 
      252     : 'Semi Expert', 
      253     : 'Semi Expert', 
      261     : 'Semi Expert', 
      262     : 'Semi Expert', 
      263     : 'Semi Expert', 
      271     : 'Semi Expert', 
      272     : 'Semi Expert', 
      281     : 'Semi Expert', 
      282     : 'Semi Expert', 
      283     : 'Semi Expert', 
      291     : 'Semi Expert', 
      292     : 'Semi Expert', 
      293     : 'Semi Expert', 
      311     : 'Officer', 
      312     : 'Officer', 
      313     : 'Officer', 
      314     : 'Officer', 
      315     : 'Officer', 
      316     : 'Officer', 
      317     : 'Officer', 
      318     : 'Officer', 
      321     : 'Officer', 
      322     : 'Officer', 
      323     : 'Officer', 
      411     : 'Service', 
      412     : 'Service', 
      413     : 'Service', 
      414     : 'Service', 
      415     : 'Service', 
      416     : 'Service', 
      421     : 'Service', 
      422     : 'Service', 
      431     : 'Service', 
      432     : 'Service', 
      441     : 'Service', 
      442     : 'Service', 
      443     : 'Service', 
      444     : 'Service', 
      511     : 'Sale', 
      512     : 'Sale', 
      513     : 'Sale', 
      521     : 'Sale', 
      522     : 'Sale', 
      530     : 'Sale', 
      611     : 'Farmer', 
      612     : 'Farmer', 
      613     : 'Farmer', 
      614     : 'Farmer', 
      615     : 'Farmer', 
      616     : 'Farmer', 
      617     : 'Farmer', 
      618     : 'Farmer', 
      620     : 'Farmer', 
      630     : 'Farmer', 
      711     : 'Skill', 
      712     : 'Skill', 
      713     : 'Skill', 
      714     : 'Skill', 
      721     : 'Skill', 
      722     : 'Skill', 
      731     : 'Skill', 
      732     : 'Skill', 
      733     : 'Skill', 
      741     : 'Skill', 
      742     : 'Skill', 
      743     : 'Skill', 
      744     : 'Skill', 
      751     : 'Skill', 
      752     : 'Skill', 
      753     : 'Skill', 
      754     : 'Skill', 
      811     : 'Device', 
      812     : 'Device', 
      813     : 'Device', 
      814     : 'Device', 
      815     : 'Device', 
      816     : 'Device', 
      817     : 'Device', 
      821     : 'Device', 
      822     : 'Device', 
      823     : 'Device', 
      824     : 'Device', 
      825     : 'Device', 
      826     : 'Device', 
      827     : 'Device', 
      828     : 'Device', 
      829     : 'Device', 
      831     : 'Device', 
      832     : 'Device', 
      833     : 'Device', 
      841     : 'Device', 
      842     : 'Device', 
      843     : 'Device', 
      844     : 'Device', 
      911     : 'Work', 
      912     : 'Work', 
      913     : 'Work', 
      914     : 'Work', 
      915     : 'Work', 
      920     : 'Work', 
      930     : 'Work', 
      941     : 'Work', 
      942     : 'Work', 
      np.nan  : 'Unknown',
      -1      : 'Unknown',
      # -1 : np.nan, 
      }
  df.work_kind = df.work_kind.map(work_kind_map)
  df.work_kind.fillna('Unknown', inplace=True)

  ############################################################################
  # work_class : 종사상 지위
  work_class_map = {
      1  : 'Regular Worker', 
      2  : 'Temporary Worker', 
      3  : 'Day Worker', 
      np.nan  : 'Unknown',
      -1      : 'Unknown',
      }
  df.work_class = df.work_class.map(work_class_map)

  ############################################################################
  # comp_type : 기업형태
  comp_type_map = {
      1  : 'Private Company', 
      2  : 'Foreign Company',
      3  : 'Public Company',
      4  : 'Corporation',
      5  : 'Government',
      6  : 'Freelancer',
      7  : 'Non-Profit',
      8  : 'ETC',
      np.nan  : 'Unknown',
      -1      : 'Unknown',
      # -1 : np.nan, 
      }
  df.comp_type = df.comp_type.map(comp_type_map)

  ############################################################################
  # contract_term : 근로 계약 기간
  contract_term_map = {
      1       : 'Limited', 
      2       : 'Unlimited',
      3       : 'Unknown',
      # 3       : np.nan, 
      np.nan  : 'Unknown',
      -1      : 'Unknown',
      # -1 : np.nan, 
      }
  df.contract_term = df.contract_term.map(contract_term_map)

  ############################################################################
  # contract_paper : 근로계약서 수령 여부
  contract_paper_map = {
      1       : 'Received', 
      2       : 'Not Received',
      np.nan  : 'Unknown',
      -1      : 'Unknown',
      # -1 : np.nan, 
      }
  df.contract_paper = df.contract_paper.map(contract_paper_map)

  ############################################################################
  # time_regular : 근로시간 규칙성
  time_regular_map = {
      1       : 'Regular', 
      2       : 'Irregular',
      3       : 'Irregular',
      np.nan  : 'Unknown',
      -1      : 'Unknown',
      # -1 : np.nan, 
      }
  df.time_regular = df.time_regular.map(time_regular_map)

  ############################################################################
  # time_is : 정규근로시간여부
  time_is_map = {
      1       : 'Yes', 
      2       : 'No',
      np.nan  : 'Unknown',
      -1      : 'Unknown',
      }
  df.time_is = df.time_is.map(time_is_map)

  ############################################################################
  # time_week_offi : 주당 정규근로시간
  df.time_week_offi.fillna(df.time_week_offi.mean(), inplace=True)
  filt_minus = df.time_week_offi < 0
  df.loc[df[filt_minus].index, 'time_week_offi'] = df.time_week_offi.mean() # 음수는 평균으로 대체

  ############################################################################
  # time_over : 초과근로 존재 여부
  time_over_map = {
      1       : 'Yes', 
      2       : 'No',
      np.nan  : 'Unknown',
      -1      : 'Unknown',
      # -1 : np.nan, 
      }
  df.time_over = df.time_over.map(time_over_map)

  ############################################################################
  # income_when : 임금지급 시기
  income_when_map = {
      1       : 'Month', 
      2       : 'Week',
      3       : 'Day',
      4       : 'Whenever',
      5       : 'Unknown',
      np.nan  : 'Unknown',
      -1      : 'Unknown',
      # -1 : np.nan, 
      }
  df.income_when = df.income_when.map(income_when_map)

  ############################################################################
  # income_type : 임금 결정 방식
  income_type_map = {
      1       : 'Year', 
      2       : 'Month',
      3       : 'Week',
      4       : 'Day',
      5       : 'Time',
      6       : 'ETC',
      7       : 'ETC',
      8       : 'ETC',
      np.nan  : 'Unknown',
      -1      : 'Unknown',
      # -1 : np.nan, 
      }
  df.income_type = df.income_type.map(income_type_map)

  ############################################################################
  # income_bonus : 성과급제 여부
  income_bonus_map = {
      1       : 'Yes', 
      2       : 'No',
      3       : 'Unknown',
      # 3       : np.nan, 
      np.nan  : 'Unknown',
      -1      : 'Unknown',
      # -1 : np.nan, 
      }
  df.income_bonus = df.income_bonus.map(income_bonus_map)

  ############################################################################
  # force : 노조 유무
  force_map = {
      1       : 'Yes', 
      2       : 'No',
      3       : 'Unknown',
      # 3       : np.nan, 
      np.nan  : 'Unknown',
      -1      : 'Unknown',
      # -1 : np.nan, 
      }
  df.force = df.force.map(force_map)

  ############################################################################
  # 직장에서 복지 제공 여부
  welfare_map = {
      1       : 'Yes', 
      2       : 'No',
      3       : 'Unknown',
      # 3       : np.nan, 
      np.nan  : 'Unknown',
      -1      : 'Unknown',
      # -1 : np.nan, 
      }

  df.wel_1    = df.wel_1.map(welfare_map)     # 법정퇴직금
  df.wel_2    = df.wel_2.map(welfare_map)     # 유급휴가
  df.wel_3    = df.wel_3.map(welfare_map)     # 생리휴가
  df.wel_4    = df.wel_4.map(welfare_map)     # 출산전후휴가
  df.wel_5    = df.wel_5.map(welfare_map)     # 육아휴직
  df.wel_6    = df.wel_6.map(welfare_map)     # 식사비용보조
  df.wel_7    = df.wel_7.map(welfare_map)     # 학비 보조
  df.wel_8    = df.wel_8.map(welfare_map)     # 주택마련 지원
  df.wel_9    = df.wel_9.map(welfare_map)     # 사내근로복지기금지원
  df.wel_10   = df.wel_10.map(welfare_map)    # 경조사 지원
  df.wel_11   = df.wel_11.map(welfare_map)    # 휴양비용지원
  df.wel_12   = df.wel_12.map(welfare_map)    # 보육비지원
  df.wel_13   = df.wel_13.map(welfare_map)    # 저축장려금지원
  df.wel_14   = df.wel_14.map(welfare_map)    # 종업원지주제도지원
  df.wel_15   = df.wel_15.map(welfare_map)    # 개인연금보험료지원
  df.wel_16   = df.wel_16.map(welfare_map)    # 생명보험보험료지원
  df.wel_17   = df.wel_17.map(welfare_map)    # 개인의료/상해보험료 지원

  ############################################################################
  # 직무만족도
  jobin_map = {
      1       : 'Very Good', 
      2       : 'Good',
      3       : 'Normal',
      4       : 'Bad',
      5       : 'Very Bad',
      np.nan  : 'Normal',
      -1      : 'Normal',
      # -1 : np.nan, 
      }
  df.jobin_1    = df.jobin_1.map(jobin_map)    # 1.임금또는보수
  df.jobin_2    = df.jobin_2.map(jobin_map)    # 2.취업의안정성
  df.jobin_3    = df.jobin_3.map(jobin_map)    # 3.하고있는일의내용
  df.jobin_4    = df.jobin_4.map(jobin_map)    # 4.근무환경
  df.jobin_5    = df.jobin_5.map(jobin_map)    # 5.근로시간
  df.jobin_6    = df.jobin_6.map(jobin_map)    # 6.개인의발전가능성
  df.jobin_7    = df.jobin_7.map(jobin_map)    # 7.의사소통및인간관계
  df.jobin_8    = df.jobin_8.map(jobin_map)    # 8.인사고과의공정성
  df.jobin_9    = df.jobin_9.map(jobin_map)    # 9.복지후생제도
  df.jobin_all  = df.jobin_all.map(jobin_map)  # 전반적일자리만족도

  ############################################################################
  # 기술, 교육 적합도
  skill_map = {
      1       : 'Very Low',
      2       : 'Low',
      3       : 'Suit',
      4       : 'High',
      5       : 'Very High',
      np.nan  : 'Suit',
      -1      : 'Suit',
      # -1 : np.nan, 
      }
  df['skill_1'] = df['skill_1'].map(skill_map) # 현재하는일교육수준적합도
  df['skill_2'] = df['skill_2'].map(skill_map) # 현재하는일기술수준적합도

  ############################################################################
  # skill_3 : 기술지식의 유용성
  skill_3_map = {
      1       : 'Very Useful', 
      2       : 'Useful',
      3       : 'Useless',
      4       : 'Very Useless',
      np.nan  : 'Useless',
      -1      : 'Useless',
      # -1 : np.nan, 
      }
  df['skill_3'] = df['skill_3'].map(skill_3_map)

  ############################################################################
  # train : 직업훈련경험
  train_map = {
      1       : 'Did Train', 
      2       : 'Training',
      3       : 'Not Train',
      np.nan  : 'Unknown',
      -1      : 'Unknown',
      # -1 : np.nan, 
      }
  df.train = df.train.map(train_map)

  ############################################################################
  # certificate : 자격증보유여부
  certificate_map = {
      1       : 'Yes Certificate', 
      2       : 'No Certificate', 
      np.nan  : 'Unknown',
      -1      : 'Unknown',
      # -1 : np.nan, 
      }
  df.certificate = df.certificate.map(certificate_map)

  ############################################################################
  # marriage : 혼인상태
  mar_map = {
      1  : 'Single', 
      2  : 'Married',
      3  : 'Separation, Divorce, Died',
      4  : 'Separation, Divorce, Died',
      5  : 'Separation, Divorce, Died',
      }
  df.marriage = df.marriage.map(mar_map)

  ############################################################################
  # 생활만족도
  satisfy_map = {
      1       : 'Very Good', 
      2       : 'Good',
      3       : 'Normal',
      4       : 'Bad',
      5       : 'Very Bad',
      np.nan  : 'Normal',
      -1      : 'Normal',
      # -1 : np.nan, 
      }
  df.satisfy_1    = df.satisfy_1.map(satisfy_map)    # 1.가족의 수입
  df.satisfy_2    = df.satisfy_2.map(satisfy_map)    # 2.여가활동
  df.satisfy_3    = df.satisfy_3.map(satisfy_map)    # 3.주거환경
  df.satisfy_4    = df.satisfy_4.map(satisfy_map)    # 4.가족관계
  df.satisfy_5    = df.satisfy_5.map(satisfy_map)    # 5.친인척 관계
  df.satisfy_6    = df.satisfy_6.map(satisfy_map)    # 6.사회적 친분 관계
  df.satisfy_all  = df.satisfy_all.map(satisfy_map)  # 전반적생활만족도

  ############################################################################
  # religion : 종교
  religion_map = {
      1       : 'No', 
      2       : 'Yes',
      3       : 'Yes',
      4       : 'Yes',
      5       : 'Yes',
      6       : 'Yes',
      7       : 'Yes',
      8       : 'Yes',
      9       : 'Yes',
      10       : 'Yes',
      np.nan  : 'Unknown',
      -1      : 'Unknown',
      # -1 : np.nan, 
      }
  df.religion = df.religion.map(religion_map)

  ############################################################################
  # 타겟변수 이진화
  stf_bool = {
      'Very Good' : 1,
      'Good'      : 1,
      'Normal'    : 0,
      'Bad'       : 0,
      'Very Bad'  : 0,
      'Unknown'   : 0,
      np.nan      : 0,
      }
  df['satisfy'] = df['satisfy_all'].map(stf_bool)

  ############################################################################
  # 설문조사 시점 year-month-date 특성 만들기 : wave_now
  df.wave_year  = df.wave_year.astype(int)
  df.wave_year  = df.wave_year + 1997
  df.wave_month = df.wave_month.astype(int)
  df.wave_date  = df.wave_date.astype(int)
  df['wave_now']     = df['wave_year'].astype(str) + '-' + df['wave_month'].astype(str) + '-' + df['wave_date'].astype(str)
  df['wave_now']     = pd.to_datetime(df['wave_now'])

  # 취업한 시점 year-month-date 특성 만들기 : workin
  df.work_year  = df.work_year.astype(int)
  df.work_month = df.work_month.astype(int)
  df['workin']  = df['work_year'].astype(str) + '-' + df['work_month'].astype(str) + '-15'
  filt          = (df['workin'].str.startswith('-1')) | (df['workin'].str.endswith('--1'))
  df.loc[df[filt]['workin'].index, 'workin'] = np.nan
  df['workin']  = pd.to_datetime(df['workin'])

  # 일자리 경력 특성 만들기 : tenure
  df['tenure'] = df['wave_now'] - df['workin']
  df['tenure'] = df['tenure'].dt.days
  filt_minus = df['tenure'] < 0
  df.loc[df[filt_minus].index, 'tenure'] = 0  # 경력이 음수인 경우는 0으로 대체
  df['tenure'] = (df['tenure'] / 30).apply(np.ceil)

  ############################################################################
  # 관심 산업분야 추출하기
  filt_com =  (df['work_industry'] == 721) | \
              (df['work_industry'] == 722) | \
              (df['work_industry'] == 723) | \
              (df['work_industry'] == 724) | \
              (df['work_industry'] == 729)

  filt_sale = (df['work_industry'] == 521) | \
              (df['work_industry'] == 522) | \
              (df['work_industry'] == 523) | \
              (df['work_industry'] == 524) | \
              (df['work_industry'] == 525) | \
              (df['work_industry'] == 526)

  filt_bld =  (df['work_industry'] == 451) | \
              (df['work_industry'] == 452)

  df.loc[df[filt_com].index, 'work_industry']  = 'Computing Data'  # 정보처리 및 기타 컴퓨터 운영 관련업
  df.loc[df[filt_sale].index, 'work_industry'] = 'Retail'          # 소매업; 자동차 제외
  df.loc[df[filt_bld].index, 'work_industry']  = 'Building'        # 종합 건설업

  filt_industry = (df['work_industry'] == 'Computing Data') | \
                  (df['work_industry'] == 'Retail') | \
                  (df['work_industry'] == 'Building')
  df = df[filt_industry]

  ############################################################################
  # 불필요한 변수 제거
  del_columns = [
        'workin', 'wave_now', 'region', 
        'satisfy_1', 'satisfy_2', 'satisfy_3', 'satisfy_4', 'satisfy_5', 'satisfy_6', 'satisfy_all', 
        'jobin_all', 
        ]
  df.drop(columns=del_columns, inplace=True)

  return df

df = feature_engineering(df_na);
```
